```python
plt.figure(figsize=(10, 5))
sns.countplot(x='quality', data=new_)
plt.title("Distribution of target class")
```

```python
f, axs = plt.subplots(1, 2, figsize=(20, 5), width_ratios=[3, 1])

sns.heatmap(data=new_.corr(method="pearson", numeric_only=True), vmin=-1, vmax=1, annot=True, ax=axs[0])
axs[0].set_title("Correlation matrix")

sns.heatmap(data=new_.corr(method="pearson", numeric_only=True)[["quality"]].sort_values('quality', ascending=False), vmin=-1, vmax=1, annot=True, ax=axs[1])
axs[1].set_title("Correlation of numerical attributes with Rings")
```
